```python
import numpy as np
import pandas as pd
```

```python
import plotly.graph_objects as go
```

```python
import plotly.express as px
```

```python
data=pd.read_csv("covid_19_data.csv")
```

```python
data.head()
```

```python
data=data.rename(columns={"ObservationDate":"Date","Country/Region":"Country"})
```

```python
data_1=data.groupby(['Country','Date']).sum().sort_values("Date",ascending=False).reset_index().copy()
data_1=data_1.drop_duplicates(subset=['Country'])
```

```python
data_1
```

```python
fig_1=go.Figure(data=go.Choropleth(
    z=data_1['Confirmed'],
    locationmode='country names',
    locations=data_1['Country'],
    colorscale="Reds"
)
)
fig_1
```

```python
fig_2=go.Figure(data=go.Choropleth(
    z=data_1['Deaths'],
    locationmode='country names',
    locations=data_1['Country'],
    colorscale="blackbody"
)
)
fig_2
```

```python
fig_3=go.Figure(data=go.Choropleth(
    z=data_1['Recovered'],
    locationmode='country names',
    locations=data_1['Country'],
    colorscale="greens"
)
)
fig_3
```

```python

data_3=data.groupby(['Date','Country']).sum().reset_index()
```

```python
data_3
```

```python
fig_11=px.choropleth(
    data_3,
    locations='Country',
    locationmode="country names",
    color="Confirmed", 
    hover_name="Country", 
    animation_frame="Date"
)
fig_11.update_layout(
    title_text = 'Global Spread of Coronavirus',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))
fig_11.show()
```

```python
fig_12=px.bar(
    data_3,
    x="Country",
    y="Confirmed",
    animation_frame="Date"
)
fig_12.show()
```
