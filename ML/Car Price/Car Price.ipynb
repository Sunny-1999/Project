## Linear Regression model

### Declare the  inputs and the targets

```python
targets = data_pp['log price']
```

```python
inputs = data_pp.drop(['log price'], axis=1)
```

### Scale the data

```python

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
inputs_s=scaler.transform(inputs)
```

```python
inputs_s
```

### Train Test Model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(inputs_s, targets, test_size=0.2, random_state=365)
```

```python
x_train
x_test
y_train
y_test
```

### Create Linear Regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

checking the model

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('y_train')
plt.ylabel('predicted values(y_hat)')
plt.xlim(6.13)
plt.ylim(6,13)
plt.show()
```

#### the closer is the plot to 45 degree the better the model

```python
sns.distplot(y_train-y_hat)
plt.title("Residual PDF", size=18)
```

```python
reg.score(x_train, y_train)
```

# Testing

```python
yhat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,yhat_test, alpha=0.2)
plt.xlabel('y_test')
plt.ylabel('predicted values(yhat_test)')
plt.xlim(6.13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_test-yhat_test)
plt.title("Residual PDF", size=18)
```

```python
df_pf=pd.DataFrame(np.exp(yhat_test), columns=['Prediction'])
df_pf.head()
```

```python
y_test=y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target']=np.exp(y_test)
```

```python
df_pf
```

```python
df_pf["Residual"]=df_pf['Target']-df_pf['Prediction']
```

```python
df_pf
```

```python
df_pf["difference%"]=np.absolute(df_pf["Residual"]/df_pf["Target"]*100)
```
